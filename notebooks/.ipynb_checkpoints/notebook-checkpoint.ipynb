# House Price Prediction

## Install lib and Get Data

```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read data
df = pd.read_csv("../Housing.csv")
```

## Questions

> 1. How does the house area affect the price?
> 2. How does the house equipment status affect the price?

### Example. Check the first 5 values in the data

```python
df.head(10)
```

### Average House Price by Area's Acreage

```python
# Calculate the average house price based on the area
avg_price_by_stories = df.groupby('stories')['price'].mean()

# Create colors 
colors = ['blue', 'green', 'red']

# Create bar chart
plt.figure(figsize=(12, 6))
avg_price_by_stories.plot(kind='bar', color=colors[:len(avg_price_by_stories)])
plt.title('House Price according to Stories ')
plt.xlabel('Stories')
plt.ylabel('Price (USD)')
plt.grid(True)
# Save image
plt.savefig('house_price_by_stories.png')
plt.show()
```

### Average House Price by Equipment status

```python
# Calculate the average house price based on the equipments status
avg_price_by_equipment_status = df.groupby('furnishingstatus')['price'].mean()

# Create colors 
colors = ['blue', 'green', 'red']

# Create bar chart
plt.figure(figsize=(12, 6))
avg_price_by_equipment_status.plot(kind='bar', color=colors[:len(avg_price_by_equipment_status)])
plt.title('Average House Price by equipments status')
plt.xlabel('Equipments status')
plt.ylabel('Average House Price')
plt.grid(True)
plt.show()
```

### Average House Price by Number of Floors

```python
# Calculate the average house price based on the number of stories
avg_price_by_floors = df.groupby('stories')['price'].mean()

# Draw a line chart
plt.figure(figsize=(12, 6))
plt.plot(avg_price_by_floors.index, avg_price_by_floors.values, marker='o', color='lightgreen', linestyle='-')
plt.title('Average House Price by Number of Floors')
plt.xlabel('Number of Floors')
plt.ylabel('Average House Price')
plt.grid(True)
plt.show()
```
