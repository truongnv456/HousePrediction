# House Price Prediction

## Install lib and Get Data

```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read data
df = pd.read_csv("../Housing.csv")
```

## Questions

> 1. How does the house pred area affect the price?<br>
> 3. How does the house equipment status affect the price?

### 1. Average House Price by pred area

```python
# Calculate the average house price by pred_area
avg_price_by_pred_area = df.groupby('prefarea')['price'].mean()

# Create colors 
colors = ['blue', 'green', 'red']

# Create bar chart
plt.figure(figsize=(12, 6))
avg_price_by_pred_area.plot(kind='bar', color=colors[:len(avg_price_by_pred_area)])
plt.title('House Price according to Prefarea ')
plt.xlabel('PredArea')
plt.ylabel('Price (USD)')
plt.grid(True)
# Save image
plt.savefig('result/house_price_by_prefarea.png')
plt.show()
```

### 2. Average House Price by Equipment status

```python
# Calculate the average house price based on the equipments status
avg_price_by_equipment_status = df.groupby('furnishingstatus')['price'].mean()

# Create colors 
colors = ['blue', 'green', 'red']

# Create bar chart
plt.figure(figsize=(12, 6))
avg_price_by_equipment_status.plot(kind='bar', color=colors[:len(avg_price_by_equipment_status)])
plt.title('Average House Price by equipments status')
plt.xlabel('Equipments status')
plt.ylabel('Average House Price')
plt.grid(True)
# Save image
plt.savefig('result/house_price_by_equipments_status.png')
plt.show()
```

### 3. Average House Price by Number of Floors

```python
# Calculate the average house price based on the number of stories
avg_price_by_floors = df.groupby('stories')['price'].mean()

# Draw a line chart
plt.figure(figsize=(12, 6))
plt.plot(avg_price_by_floors.index, avg_price_by_floors.values, marker='o', color='lightgreen', linestyle='-')
plt.title('Average House Price by Number of Floors')
plt.xlabel('Number of Floors')
plt.ylabel('Average House Price')
plt.grid(True)
# Save image
plt.savefig('result/house_price_by_stories.png')
plt.show()
```
